## データクリーニングと前処理

### Subtask:
特定された課題に基づいて、データのクリーニングと前処理を行います。具体的には、欠損値の処理（補完、削除）、異常値の特定と対応、データ型の変換、カテゴリカルデータのエンコーディングなどを行います。この際、「データが汚かったため、このように処理して整合性を保った」という実務的な記述をMarkdown形式で詳細に記録します。特に、複数のデータを結合する場合は、「なぜこのキー（ID）で結合したのか」「結合によって消えてしまったデータ（欠落）はないか」を明確に説明します。


**Reasoning**:
Based on the initial data exploration, the most pressing issues are the missing values in `Product_Category_2` and `Product_Category_3`, and the incorrect data types for `Age` and `Stay_In_Current_City_Years`. This step will address these issues by imputing missing values and converting data types to a more suitable format for analysis.



````python
import numpy as np

# 欠損値の補完（Product_Category_2, Product_Category_3）
# 欠損値の多さから、0を埋めることで「カテゴリなし」を表現
df_train['Product_Category_2'].fillna(0, inplace=True)
df_train['Product_Category_3'].fillna(0, inplace=True)
df_test['Product_Category_2'].fillna(0, inplace=True)
df_test['Product_Category_3'].fillna(0, inplace=True)

# 'Stay_In_Current_City_Years'の変換
# '4+'を4に変換し、整数型にする
df_train['Stay_In_Current_City_Years'] = df_train['Stay_In_Current_City_Years'].replace('4+', '4').astype(int)
df_test['Stay_In_Current_City_Years'] = df_test['Stay_In_Current_City_Years'].replace('4+', '4').astype(int)

# 'Age'の変換
# 年齢範囲を代表する数値（中間値）に変換する
age_mapping = {
    '0-17': 8,
    '18-25': 21,
    '26-35': 30,
    '36-45': 40,
    '46-50': 48,
    '51-55': 53,
    '55+': 60  # 55歳以上なので、代表値として60を設定
}
df_train['Age'] = df_train['Age'].map(age_mapping)
df_test['Age'] = df_test['Age'].map(age_mapping)

print("--- df_train after cleaning ---")
print(df_train.head())
print("\n--- df_test after cleaning ---")
print(df_test.head())

print("\nData cleaning and preprocessing for missing values and data types completed.")
````

**Reasoning**:
I need to encode the categorical features ('Gender', 'City_Category', 'Product_ID') and ensure that product category columns are integers. For 'Product_ID', given its high cardinality, I will initially use one-hot encoding. For 'Gender' and 'City_Category', one-hot encoding is appropriate. Finally, I'll convert the product category columns to integer type as they represent discrete categories.



````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Combine train and test for consistent encoding
# Drop 'Purchase' from df_train for combination. Using errors='ignore' in case 'Purchase' was already dropped in a previous partial execution.
df_combined = pd.concat([df_train.drop('Purchase', axis=1, errors='ignore'), df_test], ignore_index=True)

# One-hot encode 'Gender' and 'City_Category' if they exist in df_combined
cols_to_onehot_encode = []
if 'Gender' in df_combined.columns:
    cols_to_onehot_encode.append('Gender')
if 'City_Category' in df_combined.columns:
    cols_to_onehot_encode.append('City_Category')

if cols_to_onehot_encode: # Only run get_dummies if there are columns to encode
    df_combined = pd.get_dummies(df_combined, columns=cols_to_onehot_encode, drop_first=True)

# Convert Product_Category columns to integer type (they are already imputed with 0)
df_combined['Product_Category_1'] = df_combined['Product_Category_1'].astype(int)
df_combined['Product_Category_2'] = df_combined['Product_Category_2'].astype(int)
df_combined['Product_Category_3'] = df_combined['Product_Category_3'].astype(int)

# For Product_ID, given its high cardinality, label encoding is used to convert it to a numerical representation.
# This is done to avoid creating too many columns with one-hot encoding for a high-cardinality feature.
le = LabelEncoder()
df_combined['Product_ID'] = le.fit_transform(df_combined['Product_ID'])

# Separate back into df_train and df_test, ensuring copies to avoid SettingWithCopyWarning
df_train_encoded = df_combined.iloc[:len(df_train)].copy()
df_test_encoded = df_combined.iloc[len(df_train):].copy()

# Add 'Purchase' back to df_train_encoded using .loc for explicit assignment
# Only assign if 'Purchase' was originally present in df_train to avoid errors if df_train itself is already processed.
if 'Purchase' in df_train.columns:
    df_train_encoded.loc[:, 'Purchase'] = df_train['Purchase'].values

# Update the original df_train and df_test DataFrames with the encoded versions
df_train = df_train_encoded.copy()
df_test = df_test_encoded.copy()

print("-- - df_train after encoding ---")
print(df_train.head())
print("\n-- - df_test after encoding ---")
print(df_test.head())

print("\nCategorical data encoding completed.")
````

これらのエンコーディング処理は、訓練データとテストデータを結合した上で実行し、その後元のデータセットに分割することで、両データセット間で一貫したエンコーディングが適用され、データの整合性が保証されました。

## 製品カテゴリ1と購買金額の統計的検証

### Subtask:
製品カテゴリ1間の平均購買金額に統計的に有意な差があるかをANOVA検定で確認します。


**Reasoning**:
To statistically verify the observed differences in average purchase amounts across different 'Product_Category_1' groups, I will perform an ANOVA test as outlined in the instructions.



````python
from scipy.stats import f_oneway

# Group purchase amounts by Product_Category_1
product_category_1_groups = df_train.groupby('Product_Category_1')['Purchase'].apply(list)

# Perform ANOVA test
alpha = 0.05
f_statistic, p_value = f_oneway(*product_category_1_groups)

print(f"ANOVA F-statistic: {f_statistic:.2f}")
print(f"ANOVA P-value: {p_value:.3e}")

if p_value < alpha:
    print(f"P-value ({p_value:.3e}) is less than alpha ({alpha}), so we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in average purchase amounts among different Product_Category_1 groups.")
else:
    print(f"P-value ({p_value:.3e}) is greater than or equal to alpha ({alpha}), so we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in average purchase amounts among different Product_Category_1 groups.")
````
